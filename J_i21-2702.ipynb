```python
def intialization(Monday_blank_dataframe,Tuesday_blank_dataframe,courses_tuple):
    Monday_new_dataframe=copy.deepcopy(Monday_blank_dataframe)
    Tuesday_new_dataframe=copy.deepcopy(Tuesday_blank_dataframe)
    #randomly selecting null position and filling with each course
    for course in courses_tuple:
        #print(course)
        #choosing between Monday and Tuesday
        k = random.randint(0, 1)
        if(k==0):
            dataframe_to_edit=Monday_new_dataframe
        else:
            dataframe_to_edit=Tuesday_new_dataframe
        rows,cols = np.where(dataframe_to_edit.isnull())
        nan_indices = list(zip(rows, cols))
        if ((len(nan_indices)) != 0 ):     # check if nan values are available
            random_nan=random.choice(nan_indices)
            #print(random_nan)
            dataframe_to_edit.iloc[random_nan[0],random_nan[1]]=course
    return (Monday_new_dataframe,Tuesday_new_dataframe)


def section_class_frequency_on_one_day(table):
    highest_freq = 0
    section_list = []
    freq_dict = {}
    for row in table.itertuples(index=False):
        non_null_values = [val for val in row if pd.notnull(val)]
        for value in non_null_values:
            section_list = section_list + [value[2]]
    for item in section_list:
        if item in freq_dict:
            freq_dict[item] += 1
        else:
            freq_dict[item] = 1
    try:
        highest_freq = max(freq_dict.values())
        if highest_freq > 4:
            return highest_freq 
    except ValueError:
        None
    return 0



def section_class_an_same_time(table):
    count = 0
    for column_name, column in table.iteritems():
        non_null_values = [val for val in column if pd.notnull(val)]
        sections = []
        for val in non_null_values:
            sections.append(val[2])
        unique_sections = set(sections)
        if len(sections) == len(unique_sections):
            count += 1
    return count


def instructors_class_on_same_time(table):
    count = 0
    for column_name, column in table.iteritems():
        non_null_values = [val for val in column if pd.notnull(val)]
        instructors = []
        for val in non_null_values:
            instructors.append(val[0])
        unique_instructors = set(instructors)
        if len(instructors) == len(unique_instructors):
            count += 1
    return count

def teacher_classes_frequency_on_one_day(table):
    highest_freq = 0
    teacher_list = []
    freq_dict = {}
    for row in table.itertuples(index=False):
        non_null_values = [val for val in row if pd.notnull(val)]
        for value in non_null_values:
            teacher_list = teacher_list + [value[0]]
    for item in teacher_list:
        if item in freq_dict:
            freq_dict[item] += 1
        else:
            freq_dict[item] = 1
    try:
        highest_freq = max(freq_dict.values())
        if highest_freq > 3:
            return highest_freq 
    except ValueError:
        None
    return 0
    
def section_class_frequency_on_one_day(table):
    highest_freq = 0
    section_list = []
    freq_dict = {}
    for row in table.itertuples(index=False):
        non_null_values = [val for val in row if pd.notnull(val)]
        for value in non_null_values:
            section_list = section_list + [value[2]]
    for item in section_list:
        if item in freq_dict:
            freq_dict[item] += 1
        else:
            freq_dict[item] = 1
    try:
        highest_freq = max(freq_dict.values())
        if highest_freq > 4:
            return highest_freq 
    except ValueError:
        None
    return 0
    
def crossover(table_tuple1, table_tuple2):
    table1_matrix = table_tuple1[0].copy()
    table1_timetable = table_tuple1[1].copy()
    table2_matrix = table_tuple2[0].copy()
    table2_timetable = table_tuple2[1].copy()
    
    new_table_matrix = pd.concat([table1_matrix.iloc[:,:3],table2_matrix.iloc[:,3:]],axis=1)
    new_table_timetable = pd.concat([table1_timetable.iloc[:,:3],table2_timetable.iloc[:,3:]],axis=1)

    return (new_table_matrix, new_table_timetable)

    
def fitness_function(table):
    fitness_value=0
    
    if(instructors_class_on_same_time(table[0]) > 0 and instructors_class_on_same_time(table[1]) > 0):
        fitness_value+=((instructors_class_on_same_time(table[0])*5))
        #print(instructors_class_on_same_time(table[1]))
        #print(instructors_class_on_same_time(table[1]))
#     else:
#         print(1)
        
    if(section_class_an_same_time(table[0]) > 0 and section_class_an_same_time(table[1]) > 0 ):
        fitness_value+=((section_class_an_same_time(table[0])*5 ))
#     else:
#         print(2)
    
    if (teacher_classes_frequency_on_one_day(table[0]) == 0):
        fitness_value+=7.5
    else:
        fitness_value+=(7.5 - (teacher_classes_frequency_on_one_day(table[0]))/2)
        
    if (teacher_classes_frequency_on_one_day(table[1]) == 0):
        fitness_value+=7.5
    else:
        fitness_value+=(7.5 - (teacher_classes_frequency_on_one_day(table[1]))/2)
        
    
    if (section_class_frequency_on_one_day(table[0]) == 0):
        fitness_value+=7.5
    else:
        fitness_value+=(7.5 - (section_class_frequency_on_one_day(table[0]) )/2)
    
    if (section_class_frequency_on_one_day(table[1]) == 0):
        fitness_value+=7.5
    else:
        fitness_value+=(7.5 - (section_class_frequency_on_one_day(table[0]) )/2)
        
    return fitness_value


    
import pandas as pd

def count_column_no(df,col_name):
    count=0
    for i in df.columns:
        if(i==col_name):
            return count
        
    return -1
def mutation(table_tuple):
    tables = [table_tuple[0].copy(), table_tuple[1].copy()]

    for table in tables:
        # Check for instructors scheduled at the same time
        for column_name, column in table.iteritems():
            non_null_values = [val for val in column if pd.notnull(val)]
            instructures = [instructure[2] for instructure in non_null_values]
            unique_instructors = set(instructures)
            if len(instructures) != len(unique_instructors):
                for instructor in unique_instructors:
                    count = 0
                    for i, instructure in enumerate(instructures):
                        if instructure == instructor:
                            count += 1
                            if count >= 2:
                                #print(count_column_no(table,column_name))
                                col_index = count_column_no(table,column_name)
                                #row2 = column.index[i+1] # i+1 is the row no ie class C-301 etc
                                #print(row2)
                                #print(table)
                                temp = table.iloc[i+1, col_index]
                                table.iloc[i+1, col_index] = table.iloc[i, col_index]
                                table.iloc[i, col_index] = temp
                                break
        
        # Check for sections scheduled at the same time
        for column_name, column in table.iteritems():
            non_null_values = [val for val in column if pd.notnull(val)]
            sections = [instructure[1] for instructure in non_null_values]
            unique_sections = set(sections)
            if len(sections) != len(unique_sections):
                for section in unique_sections:
                    count = 0
                    for i, instructure in enumerate(sections):
                        if instructure == section:
                            count += 1
                            if count >= 2:
                                col_index = count_column_no(table,column_name)
                                temp = table.iloc[i+1, col_index]
                                table.iloc[i+1, col_index] = table.iloc[i, col_index]
                                table.iloc[i, col_index] = temp
                                break

    return (tables[0], tables[1])

        

def tournament(tournament_list):
    while(len(tournament_list)>1):
        new_list=[]
        for i in range(int(len(tournament_list)/2)):
            new_list.append(max(tournament_list[i*2],tournament_list[(i*2)+1]))
        
        tournament_list=new_list
    return tournament_list[0]

def genetic_algorithm(Monday_blank_dataframe,Tuesday_blank_dataframe,courses_tuple,population_number):
    chromosome_fitness_tuples_list=[]
    fitnesses=[]
    #intializing
    for population in range(population_number):
        chromosome=intialization(Monday_blank_dataframe,Tuesday_blank_dataframe,courses_tuple)
#         chromosomes.append(chromosome)
        fitness=fitness_function(chromosome)
#         print(f"fitness of {population+1} :", fitness)
        chromosome_fitness_tuples_list.append((chromosome,fitness))
    
    try_=0
    
    while(1):
        if(try_%3==0):
                for population in range(population_number):
                    chromosome=intialization(Monday_blank_dataframe,Tuesday_blank_dataframe,courses_tuple)
                    fitness=fitness_function(chromosome)
#                     print(f"fitness of {population+1} :", fitness)
                    chromosome_fitness_tuples_list.append((chromosome,fitness))

        if(len(chromosome_fitness_tuples_list)>population_number):
            for position in range(len(chromosome_fitness_tuples_list)):
#                 print(chromosome_fitness_tuples_list[position][0])
#                 print(position)
                chromosome_fitness_tuples_list[position]=(chromosome_fitness_tuples_list[position][0],fitness_function(chromosome_fitness_tuples_list[position][0]))
        
            chromosome_fitness_tuples_list = sorted(chromosome_fitness_tuples_list, key=lambda x: x[1], reverse=True)
            chromosome_fitness_tuples_list=chromosome_fitness_tuples_list[:population_number]
                
                        
                        
        for chromosome_fitness_tuples in chromosome_fitness_tuples_list:
            if chromosome_fitness_tuples[1] > 80:
                print(chromosome_fitness_tuples[1])
                print("found")
                
                print("MONDAY")
                display(chromosome_fitness_tuples[0][0])
                print("TUESDAY")
                display(chromosome_fitness_tuples[0][1])
                
                return
        
        random.shuffle(chromosome_fitness_tuples_list)
        fitnesses = [chromosome_fitness_tuples[1] for chromosome_fitness_tuples in chromosome_fitness_tuples_list]
        print(fitnesses)
        
        #tournament
        tournament_values = []
        for i in range(population_number):
            value = tournament(fitnesses)
            tournament_values.append(value)
            fitnesses.remove(value)
    
    
        
#         print(chromosome_fitness_tuples_list)

        tournament_tuple_values = []
        for i in range(population_number):
            tuple_value = next((chromosome_fitness_tuples for chromosome_fitness_tuples in chromosome_fitness_tuples_list if chromosome_fitness_tuples[1] == tournament_values[i]), None)
            tournament_tuple_values.append(tuple_value)
    
              
        
        #crossover
        fetuses=[]
        count=0
        for i in range(population_number-1):
            fetus=crossover(tournament_tuple_values[i][0],tournament_tuple_values[i+1][0])
            fetuses.append(fetus)
            i=i+1
        for i in range(population_number-1):
            fetus=crossover(tournament_tuple_values[i+1][0],tournament_tuple_values[i][0])
            fetuses.append(fetus)
            i=i+1
        

        for i in range(population_number):
            chromosome_fitness_tuples_list.append((fetuses[i],0))
        
#         print(fetus1)
        
        #mutation
        mutations=[]
        for i in range(population_number):
            mutation_=mutation(tournament_tuple_values[i][0])
            mutations.append(mutation_)
        
        #         print(mutation1)
        
        for i in range(population_number):
            chromosome_fitness_tuples_list.append((mutations[i],0))
        #print(chromosome_fitness_tuples_list[0])
       
```
